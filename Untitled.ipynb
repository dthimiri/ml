# Supervised Learning


## Classification and Regression

There are 2 major types of supervised machine learning problems:
1.) Classification
2.) Regression

### Classification:

1.) Goal is to predict a class label 

2.) Classification is sometimes separated into binary classification, which is a special case of distincguishing exactly 2 classes and multiclass classification which is classification between 2 or more classes.

3.) In case of Regression, the goal is predict a continous variable.

```python
# Breast Cancer Dataset
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

cancer = load_breast_cancer()

# Data description
print("cancer.keys(): \n{}" .format(cancer.keys()))
print("\ncancer.data.shape{}" .format(cancer.data.shape))
print("\nSample counts per class: \n{}" .format({x: y for x, y in zip(cancer.target_names, np.bincount(cancer.target))}))
print("\nFeature names: \n{}".format(cancer.feature_names))

# create the train and test data
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=66)

training_accuracy = []
test_accuracy = []

# n_neighbor range from 1 to 10
neighbors_settings = range(1, 10)

for n_neighbors in neighbors_settings:
    
    # Instantiate the KNeighborsClassifier: knn
    knn = KNeighborsClassifier(n_neighbors=n_neighbors)
    # fit the training data
    knn.fit(X_train, y_train)
    training_accuracy.append(knn.score(X_train, y_train))
    test_accuracy.append(knn.score(X_test, y_test))
    
#plt.figure()
plt.plot(neighbors_settings, training_accuracy, label="training accuracy")
plt.plot(neighbors_settings, test_accuracy, label="test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.legend()
plt.show()
```

```python
# Analyzing KNeighbors Regressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression

# Data
X, y = mglearn.datasets.make_wave(n_samples=60)

# training and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)


lr = LinearRegression().fit(X_train, y_train)
print("Liner Regression:")
print('*'* len("linear regression"))
print("lr.coef_: {}" .format(lr.coef_))
print("lr.intercept_: {}" .format(lr.intercept_))
print("Training set score: {:.2f}" .format(lr.score(X_train, y_train)))
print("Test set score: {:.2f}" .format(lr.score(X_test, y_test)))
```

```python
# Ridge Regression

X, y =mglearn.datasets.load_extended_boston()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
lr = LinearRegression().fit(X_train, y_train)

print("Training accuract: {}" .format(lr.score(X_train, y_train)))
print("Test accuract: {}" .format(lr.score(X_test, y_test)))

print("This discrepancy is due to overfitting the training data. Hence we need to fit a model that allows to control complexity.")

from sklearn.linear_model import Ridge
ridge = Ridge().fit(X_train, y_train)
print("\nRidge Regression:")
print('*'* len("ridge regression"))

print("Training accuract: {}" .format(ridge.score(X_train, y_train)))
print("Test accuract: {}" .format(ridge.score(X_test, y_test)))
```
